#### Function to determine match outcome

```python
def match_outcome(r):
    home_score = r[0]
    away_score = r[1]
    home_team_name = r[2]
    away_team_name = r[3]
    
    if home_score > away_score:
        return 'Home'
    elif home_score < away_score:
        return 'Away'
    else:
        return 'Tie'

df_matches['match_outcome']  = df_matches[['home_team_goal', 'away_team_goal', 'home_team_name', 'away_team_name']].apply(match_outcome, axis=1)
```

```python
# drop some columns

df_matches.drop(['stage', 'match_api_id'], axis=1, inplace=True)
```

## Exploratory Data Analysis


### Research Question 1 (What teams improved the most over the time period in terms of goal scored?)

 I selected year 2008, and 2011 to compare the details of the teams. In order words, using these years we analyze what teams improved the most.

```python
df_matches_clean = pd.read_csv("matches_clean.csv")
df_matches_clean.head()
```

```python
# number of unique years in the table
years = df_matches_clean.year.unique()
years
```

```python
#number of away teams
len(df_matches_clean.away_team_name.unique())
```

```python
# number of home teams
len(df_matches_clean.home_team_name.unique())
```

#### Use statistical analysis to estimate

```python
df_match_2008 = df_matches_clean.query('year == 2008')

df_match_2011 = df_matches_clean.query('year == 2011')

df_match_2011_away = df_match_2011.groupby(['away_team_name'])['away_team_goal'].mean() # average away team goals

df_match_2011_home = df_match_2011.groupby(['home_team_name'])['home_team_goal'].mean() # average home team goals

df_match_2008_away = df_match_2008.groupby(['away_team_name'])['away_team_goal'].mean()

df_match_2008_home = df_match_2008.groupby(['home_team_name'])['home_team_goal'].mean()

df_match_total_2011 = (df_match_2011_away + df_match_2011_home) / 2 #average of away and home goals in 2011

df_match_total_2008 = (df_match_2008_away + df_match_2008_home) / 2 # average of away and home goals in 2008

diff_match_2011_2008 = df_match_total_2011 - df_match_total_2008
```

```python
diff_match_2011_2008
```

```python
diff_match_2011_2008.dropna(inplace=True) # drop null values
```

```python
print('Improved teams :' + str(diff_match_2011_2008[diff_match_2011_2008.sort_values() > 0].count()))
```

```python
# sort the values 
sorted_index = diff_match_2011_2008.sort_values().index
```

```python
sorted_index
```

### Plot a horizontal plot to visualize the improved teams 

```python
fig, ax = plt.subplots(figsize=(15, 25))

# plot a horizontal bar
plt.barh(range(0,len(sorted_index)), diff_match_2011_2008.sort_values())

# Set the position of the y ticks
ax.set_yticks(range(0,len(sorted_index)))

# Set the position of the y ticks labels
ax.set_yticklabels(sorted_index)

# Set the y axis label
ax.set_ylabel('Teams')

# Set the chart's title
ax.set_title('Improvements Team Quality on average goals scored from 2008 to 2011')

# Set the y axis label 
plt.xlabel("Averages")

plt.tight_layout();
```
